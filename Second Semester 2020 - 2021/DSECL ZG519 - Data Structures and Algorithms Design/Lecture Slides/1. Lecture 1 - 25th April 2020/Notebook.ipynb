```python
def fact2(n):
    if n == 0:
        return 1
    else:
        return n * fact2(n-1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(221)
ax2=fig.add_subplot(222)

ax3=fig.add_subplot(223)
ax4=fig.add_subplot(224)
n=np.arange(start=1,stop=7,step=1)
ax1.set_title('Comparison of n^2 and n^3',fontsize=12)
ax1.plot(n,(n**2),'b')
ax1.plot(n,(n**3),'r')
ax1.grid(True)

ax2.set_title('Comparison of 2^(n-1) and 2^n',fontsize=12)
ax2.plot(n,2**(n-1),'b')
ax2.plot(n,2**n,'r')

ax3.title.set_text('When argument is increased four fold')
ax3.plot(n,n**(0.5),'b')
ax3.plot(n,2*(n**(0.5)),'r')
ax3.grid(True)

n=np.arange(start=2,stop=100,step=2)
ax4.title.set_text('Log n and Ln n')
ax4.plot(n,np.log(n),'b')
ax4.grid(True)
ax4.plot(n,np.log2(n),'r')


```
